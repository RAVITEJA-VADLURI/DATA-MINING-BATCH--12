```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --------------------------------------------
# Create synthetic Heart Attack dataset
np.random.seed(42)  # for reproducibility

data = {
    'age': np.random.randint(29, 77, 300),  # Ages between 29 and 76
    'sex': np.random.choice([0, 1], 300),    # 0 = Female, 1 = Male
    'cp': np.random.randint(0, 4, 300),      # Chest pain type (0-3)
    'trestbps': np.random.randint(94, 200, 300),  # Resting BP
    'chol': np.random.randint(126, 564, 300),     # Cholesterol levels
    'fbs': np.random.choice([0, 1], 300),    # Fasting blood sugar
    'restecg': np.random.randint(0, 2, 300), # Resting ECG results
    'thalach': np.random.randint(71, 202, 300),   # Max heart rate
    'exang': np.random.choice([0, 1], 300),  # Exercise-induced angina
    'oldpeak': np.round(np.random.uniform(0.0, 6.2, 300), 1),  # ST depression
    'slope': np.random.randint(0, 3, 300),   # Slope of peak exercise ST
    'ca': np.random.randint(0, 4, 300),       # Number of major vessels
    'thal': np.random.choice([3, 6, 7], 300), # Thalassemia type
    'target': np.random.choice([0, 1], 300)   # Heart Disease (0/1)
}

df = pd.DataFrame(data)

# --------------------------------------------
# Exploratory Data Analysis (EDA) and Graphs

sns.set(style="whitegrid")

# 1. Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Features", fontsize=16)
plt.show()

# 2. Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True, color='skyblue')
plt.title("Age Distribution of Patients", fontsize=16)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# 3. Chest Pain Type vs Heart Disease
plt.figure(figsize=(8, 6))
sns.countplot(x='cp', hue='target', data=df, palette='Set2')
plt.title("Chest Pain Type vs Heart Disease", fontsize=16)
plt.xlabel("Chest Pain Type (cp)")
plt.ylabel("Count")
plt.legend(title="Heart Disease (target)")
plt.show()

# 4. Gender vs Heart Disease
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='target', data=df, palette='viridis')
plt.title("Gender vs Heart Disease", fontsize=16)
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.legend(title="Heart Disease (target)")
plt.show()
```
