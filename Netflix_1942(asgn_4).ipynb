```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/netflix_titles.csv')
print(df.describe())
```

```python
plt.hist(df['release_year'].dropna(), bins=10, edgecolor='white')
plt.xlabel('count')
plt.ylabel('release_year')
plt.title('Distribution of Age')
plt.show()
```

```python
# Bar chart for the frequency of different 'types' of shows
plt.figure(figsize=(6, 4))
df['type'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Netflix Show Types')
plt.xlabel('Show Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Box plot for 'release_year' to visualize outliers and distribution
plt.figure(figsize=(6, 4))
plt.boxplot(df['release_year'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of Release Year')
plt.ylabel('Release Year')
plt.show()
```

```python
import seaborn as sns
# Pairplot to visualize relationships between numerical features
# Replace 'count' with an actual numerical column from your DataFrame
# For example, if you have a column named 'duration', you can use:
sns.pairplot(df[['duration', 'release_year']].dropna())
plt.show()
```

```python
# Convert 'duration' column to string type before using .str.replace
df['duration'] = df['duration'].astype(str).str.replace(' min', '', regex=False)
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

# Now calculate the correlation matrix
correlation_matrix = df[['duration', 'release_year']].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Cost and release year')
plt.show()
```

```python
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Show Type and Count')
plt.xlabel('Show Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
sns.violinplot(x='type', y='release_year', data=df, palette='muted')
plt.title('Release Year Distribution by Show Type')  # Update title for clarity
plt.show()
```
